```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt

# Upload image
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Read as grayscale
img = cv2.imread(filename, 0)

# Contrast adjustment (you can change alpha)
alpha = 1.5
beta = 0

contrast_img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

# Show both images
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(contrast_img, cmap='gray')
plt.title("Contrast Adjusted")
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(contrast_img, cmap='gray')
plt.title("Contrast Adjusted")
plt.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

# Histogram Original
plt.subplot(1, 2, 1)
plt.hist(img.ravel(), bins=256, color='black')
plt.title("Histogram (Original)")

# Histogram Contrast Adjusted
plt.subplot(1, 2, 2)
plt.hist(contrast_img.ravel(), bins=256, color='black')
plt.title("Histogram (Contrast Adjusted)")

plt.show()
```
